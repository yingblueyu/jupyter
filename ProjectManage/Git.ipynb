## Add files to the staging area, untracked file.  

- 在.git/objects目录下生成blob文件
- sha1(“blob” + 内容长度 + 内容) => 160bit哈希值作为文件名  
- (“blob” + 内容长度 + 内容)被压缩(zlib)后存储--blob object  
- 只存储文件内容，无文件名，两个文件内容相同，生成一个blob文件 

```
git add filename # 生成一个blog文件, To add files to objects and index


git add -A # To add all the files you have edited or changed in your working directory.
git add -all

git cat-file -t 8d0e41 // 查看objects里面文件类型: blob
git cat-file -s 8d0e41 // 大小
git cat-file -p 8d0e41 // 查看objects里面文件内容

```

- 文件名信息存储在索引里 .git/index

```
git ls-files // 列出索引区文件
git ls-files -s  // 索引区文件权限(与实际文件类型有关), 对应的blob对象, 文件名

git status  // 工作区与索引区的区别
```

Commit: tracked file.
- A commit in Git is basically one version of a project. 
- A commit hash is 51dc6ecb327578cca503abba4a56e8c18f3835e1
- can just use 51dc6ec to refer to the commit.

```
git commit -m "comment"
git log # list the commits in local repo in reverse order
git ls-files -s // blob对象 and 文件名
```
